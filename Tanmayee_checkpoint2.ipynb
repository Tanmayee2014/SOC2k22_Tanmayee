```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from IPython.core.pylabtools import figsize
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv("/content/drive/MyDrive/my_dataset/credit_train.csv")
df.head()
```

```python
print("Shape:")
print(df.shape)
print("Columns")
print(df.columns)
print("info")
print(df.describe())
```

```python
for col in df.columns:
    print('\n'+col+"\nUnique:",df[col].nunique())
    print("Empty_values:",df[col].isnull().sum())
```

```python
df[df['Loan ID'].isnull()==True]
#from result we can see that last 514 rows are empty
```

```python
#Dropping rows having Null Loan id
df.drop(df[df['Loan ID'].isnull()==True].index,inplace=True)
#To reset index
df.reset_index(drop=True,inplace=True) 
print(df.shape)
```

```python
print("Num of Duplicate rows:",df.duplicated().sum())
print("Size before dropping:",df.shape)
df.drop_duplicates(inplace=True)
df.reset_index(drop=True,inplace=True) 
print("Size after dropping:",df.shape)
print("Num of Duplicate rows:",df.duplicated().sum())
```

```python
#print(df.duplicated().sum())
df[df['Loan ID'].duplicated()==True]
```

```python
print("Number of Empty_values and % of Total Valuese\n")
for col in df.columns:
    if df[col].isnull().sum()!=0:
        print(col+":\n",df[col].isnull().sum(),' (',round(df[col].isnull().sum()/len(df)*100,2),')%\n',sep='')
```

```python
#deliquent column has more than 50 % null values, will drop it.
df.drop("Months since last delinquent",axis=1,inplace=True) 
df.reset_index(drop=True,inplace=True) 
print("Size after dropping:",df.shape)
```

```python
print("Number of Empty_values and % of Total Valuese\n")
for col in df.columns:
    if df[col].isnull().sum()!=0:
        print(col+":\n",df[col].isnull().sum(),' (',round(df[col].isnull().sum()/len(df)*100,2),')%\n',sep='')
```

```python
#Maximum Open Credit and Tax Liens has very low number of missing values, will drop that rows
df.drop(df[df['Maximum Open Credit'].isnull() | df['Tax Liens'].isnull()].index,inplace=True)
df.reset_index(drop=True,inplace=True) 
```

```python
df[df['Annual Income'].isnull() & df['Credit Score'].isnull()]
#Rows having missing values for both Annual Income and Credit Score are same
```

```python
#Will fill mean values in place of Null values, that will be definitely better than dropping
df.fillna(df.mean(), inplace=True)
```

```python
plt.figure(figsize=(10,8))
df['Years in current job'].hist()
plt.show()
# "Years in current job" column has catagorical values
```

```python
#Will replace null values with mode i.e. 10+ years
df.fillna('10+ years', inplace=True) 
```
